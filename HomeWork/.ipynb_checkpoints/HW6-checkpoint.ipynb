### a)	Sum of Digits of a Number:

Problem: Write a program that calculates the sum of the digits of a given number. \
Input: An integer \
Process: Add the last digit (number % 10) to the sum and remove the last digit by dividing the number by 10. \
Output: The value of the sum

### b)	Sorting a List of Numbers (Bubble Sort)

Input: A list of numbers in a text file. \
Process: Compare adjacent elements and swap them if they are in the wrong order. Repeat until the list is sorted. \
Output: The sorted list (ascending order) on screen

### b) Sorting a List of Numbers (Bubble Sort)

**Flowchart:**

<img src="Assets/HW6B.jpg" width="400">

**Pseudocode:**

```
FUNCTION bubble_sort(list)
BEGIN FUNCTION
	
	SET swap TO True
	WHILE swap = True DO	
		SET index TO 0
		SET swap TO False
		REPEAT LENGTH(list) - 1 TIME
			SET tempA TO list[index]
			SET tempB TO list[index +1]
			
			IF tempA > tempB THEN
				SET swap TO True
				SET list[index +1] TO tempA
			SET list[index] TO tempB
			END IF
			SET index TO index + 1
		END REPEAT
	END WHILE
	RETURN list
END FUNCTION

#main program:
READ number.txt list
SET orderList TO bubble_sort(list)
SEND orderList TO DISPLAY


```

**Pyhton:**

````python
#Change each element of the list into integer
def str_to_int(stringList):
    
    stringList = [int(num) for num in stringList]

    return stringList

#Bubble sort algorithm
def bubble_sort(list):

    #copy of the list such to not change the original list
    sortedList = list.copy()

    #initialisation
    swap = True

    #main loop, loop is done while a swap is performed
    while swap:
        index = 0
        swap = False

        #loop over the list
        for i in range(len(list)-1):    
            tempA = sortedList[index]
            tempB = sortedList[index + 1]
 
            # compare the two adjacent valuse
            if tempA > tempB : 
                swap = True
                sortedList[index + 1] = tempA
                sortedList[index]     = tempB
            index = index + 1

    return sortedList

#same algorithm with an optimization
def bubble_sort_opti(list):

    sortedList = list.copy()

    swap = True
    
    opti = 0 
    
    while swap:
        index = 0
        swap = False
        #here is the optimisation: since we are sure that
        #after each iteration one more value is in order
        #we do not need to loop over all the elements
        for i in range(len(list) -1 - opti):    
            tempA = sortedList[index]
            tempB = sortedList[index + 1]
            if tempA > tempB : 
                swap = True
                sortedList[index + 1] = tempA
                sortedList[index]     = tempB
            index = index + 1
        
        opti = opti + 1
  
    return sortedList

#opening the file then closing ounce read. 
myFile = open('number.dat', 'r')
numberList = myFile.read().split(',')
myFile.close()

#the list from the file is a list of string
#first step is to change each element into an integer
numberList = str_to_int(numberList)

#call the bubble sort function
sortedList = bubble_sort(numberList)

#output on the screen
print(sortedList)
````
