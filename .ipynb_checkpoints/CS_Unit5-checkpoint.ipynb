## 21 Networks

- **network**: an arrangement of computers and other devices connected together to share ressources and data.
- **network service**: an app running on a server which provides facilities or operations such as data storage, printing or communications.

### Computer Networks and their Use

A **network** is a collection of two or more computers that are connected together for the purpose of sharing resources and data.
Many networks include servers. A server provides the network with **services** (storing files, emails, computation...)

#### Why are Networks Used?

A network can support multiple users accessing multiple services at the same time.

The network allows multiples users:
- to read/write files - this provides extra space and also gives a back-up facility
- to access shared files
- to download data or updates
- to print
- to access the Internet
- to communicate with each other.

### Different Types of Network

<img src="Img/U5_21_WAN.jpg" width="500">

- **Local Area Network (LAN)** a network that covers a relatively small geographical area, often a single site.
- **Wireless Local Area Network (WLAN)** a local area network in which connected devices use high-frequency radio waves to communicate.
- **Wide Area Network (WAN)** a network that covers a large geographical area. It connects together two or more LANs and is usually under shared ownership. The largest wide area network is the Internet.
- **Personal Area Network (PAN)** network used for data transmission over short distances bu computer devices, such as laptops, mobile phones, tablets, media players, speakers and printers.
- **Bluetooth™** a protocol for the short-range wireless interconnections of mobile phones, computers and other electronic devices.

The *PAN* may just connect the local devices or allow them to connect to higher-type networks such as the internet. The devices usually communicate wirelessly over a distance of up to 10 metres by radio waves using a technology called **Bluetooth™**.

### Network Topologies

- **network topology** describes how the devices on a network are connected together.

There are four main **network topology**: bus, ring, star and mesh.

#### Bus

A bus topology is a network topology where all devices are connected to a single cable, called a "bus", that serves as a shared communication medium. In this configuration, all devices on the network receive the same signal at the same time

A bus network needs a system to deal with multiple devices sending a message at the same time. The most common method is called Carrier Sense Multiple Access with Collision Detection (CSMA/CD). A coolision occurs when two or more computers want to send some data at the same time.

<img src="Img/U5_21_Bus.png" width="400">

| ---------- | -------------| 
| Relatively cheap to install since only one cable is needed | Whole network will fail if the cable is cut or damaged |
Easy to add extra network devices | Can be difficult to identify where a fault is on the cable |
 | | The more devices that are added the slower they run. Only one message can be sent at once so more collisions and the computer needs to wait longer |
 | | All data is received by all devices on the network; this is a security risk |

 #### Ring

 A ring network topology is a network design where devices are connected in a circle, forming a closed loop that allows data to travel in one direction. Each network device has what can be thought of as an 'in' and an 'out' connection. There are no collisions. Data is passed from one device to the next intil it reaches its destination.

| | Can be difficult to identify where a fault is on the network |
| | More expensive to install than a bus network as it requires more cable to complete the ring. |

#### Star

A star network topology is a computer network design where all devices are connected to a central hub, or switch, that acts as a focal point for data transmission

| A new device can be added or removed without having to close the network down|

#### Mesh

A mesh network topology is a decentralized network configuration where devices are connected directly to each other, forming a mesh-like structure. There are two main types of mesh topology - fully connected and partially connected.

The largest mesh network of all is the **Internet**.

- **Internet** a worldwide system of interconnected networks that enables information to be exchanged and shared.

### Communication Media

Networks are either wired or wireless according to the transmission or **communication media** used.

- **Communication Media** the means by which data is transmitted between devices on a network. Coaxial cable, fibre-optic cable and microwaves are all forms of communication media.

#### Wired

There is a physical connection between computers/servers. Wired connection are made of copper wire or glass/plastic for fibre-optic cable. Fibre-optic cables transmit data far faster and allow signals to travel over greater distances than copper wire.

#### Wireless

There is NO physical connection between devices. Most wireless connections transmit and receive radio signals/microwave. Infra-red light can be used over limited distances.

### Protocols

A protocol is a set of rule that control how communications between devices are formatted and how these communications will be sent/received.

A protocol might contain details of:
- how each computer will be identified (its adress)
- what route the data will take to get to its destination (routing information)
- how errors will be found and dealt with (error checking)
- whether each part of a message should be acknowledged as received correctly
- what to do if data isn't received correctly
- how the data is to be formatted
- how the data is to be sequenced
- how the speed of the sender and receiver can be **synchronised**

The main **protocols** in use relating to networks and the Internet are detailed below.

- **protocol** a set of rules that govern how communications on a network should be formatted and what data they should include.
- **synchronised** occur at the same time or operate at the same rate.

#### Network Protocols

##### Ethernet

Ethernet is a family of protocols that are used in wired LANs. They cover everything from the physical parts to the logical parts.

##### Wi-Fi

Wi-Fi is a digital communications protocol that sets out how data is transmitted on wireless LANs.

##### TCP

The Transmission Control Protocol (TCP) is a protocol that ensures reliable data transmission between devices on a network

- Specifying that the receiving computer sends aknowledgements that section of the data sent has been received.
- using **checksums** to ensure that the data received is accurate
- allowing the receiving computer to tell the sending computer to slow down transmission.
- ensuring that data sent up to the application layer contains no duplicates and is in the correct order.

TCP is used when you access web pages, send/receive email or upload/download files.

- **checksum** A checksum is a small block of data that's used to detect errors in digital data during transmission or storage. Checksums are calculated using a checksum algorithm, which processes the data to produce a fixed-size string or number. This value acts as a fingerprint for the data, allowing users to verify its integrity.

##### TCP/IP

TCP/IP stands for Transmission Control Protocol/Internet Protocol. TCP/IP is a protocol stack, which means that it is a collection of protocols that work together.
Data sent using TCP/IP is broken up into **packets**.

- **packet** a small quantity of data being sent through a network. The packet is labelled with the sender's address (source), the recipient's address (destination), how many packets are being transmitted and the position of this packet in the complete message.

### Benefits of Using Networking Layers

Networking technologies are separated into layers, each one containing specific hardware and software and software protocols.

The benefits of this approach are:

- It makes the overall model easier to understand by dividing it into functional parts
- Each layer is specialised to perform a particular function
- The different layers can be combined in different ways.
- One layer can be developed or changed without affecting the other layers.
- It makes it easier to identify and correct networking errors and problems
- It provides a universal standard for hardware and software manufactures to follow. THis is so that theu will be able to communicate with each other.

### Mobile Communications

Mobile communications is the use of portable devices to interact with others and information sources without the need for cable. Mobile communications use high-frequency electromagnetic fields to transmit data and voice wirelessly. Mobile communications handover, also known as handoff, is the process of transferring a connection from one channel to another in a mobile network. This can happen when a device moves out of the coverage area of one cell or when the network needs to balance traffic load.

The standard for mobile communications is called Global System for Mobile Communications (GSM) to describe the protocols to be used:

- 2G (Second Generation): The second generation of mobile communications, which was introduced in the 1990s. 2G networks supportedmultimedia messaging service (MMS)s and SMS.
- 3G (Third Generation): Introduced the ability to access mobile internet with moderate speeds, enabling features like video calling and basic mobile data usage.
- 4G (Fourth Generation ):Provided significantly faster data speeds compared to 3G, allowing for high-definition video streaming, online gaming, and more reliable mobile internet access. 
- 5G (Fifth Generation): The latest generation of mobile network technology, offering extremely high data transfer rates, minimal latency, and the capacity to support a large number of connected devices, enabling applications like augmented reality and autonomous driving.

## 22 Network Security

### Network Security and its Importance.

**Network Security** protect a network and its data from threats to its confidentiality, correctness (integrity) and availability.

- **Network Security**: Activities designed to protect a network and its data from threats such as viruses, hacker attacks, denial of service attacks, data interception and theft and equipment failure.

#### Confidentiality

It is important that the network is not hacked so data is not intercepted or stolen by criminals or competitors.\
Some way to protect data confidentiality include:
- ensuring only authorised users can access the parts of a network and its resources that they have reason to require.
- stopping misuse.
- encrypting data, even if unauthorised person gains access to data, they won't be able to read it.

#### Correctness

Use of the network to communicate and store data must not change the data or allow the data to be changed without authorisation.

#### Availabilty

A network should be accessible when it is needed. Complex systems are likely to fail at some time but it is also important that protection is put in place to prevent failures caused by criminals. Virus or **Denial of Service** attacks.

-  **Denial of Service (DoS)**: an attack on a network that attempts to prevent legitimate users from accessing its services.

### Reasons why Security is important

- *Required for running the organisation*: Nowadays everything use/is stored a network infrastructure.
- *Private and confidential*: There are many types of privates and confedential data that people or businesses wouldn't want to make public. (e.g. medical records)
- *financially valuable*: Many types of data stored on business networks might be financially valuable. (e.g. business contracts)

### Authentication and Validation

Authentication is the process of checking the identity of a user of a computer system or network, often using username/password. Other systems can also be used, such as PIN or fingerprint or facial...
Sometimes a two step authentication is used called **two-factor authentication**.

- **two-factor authentication** a security check where users have to type the code from a portable hardware device called a 'secure token' or from an SMS message sent to their mobile phone.

### Other Ways to Secure a Network

Authentication alone is not enough. Three others you need to know about are: **access control**, firewalls and physical security.

#### Access control

Once a user has been authenticated, **acess control** is the method that controls whether the user will gain access to a file and what the user can do (file permissions).
- read-only access; The user cannot modify or delete the file.
- read and write access; the user can modify the file.
- delete; in some network the user might need further permission to delete a file.

- **acess control** this decides which users have access to which data, and what they are allowed to do with it.


#### Firewall

A firewall is a network security device that regulates traffic between a trusted internal network and an untrusted external network. Firewalls are an essential part of any security architecture and are an organization's first line of defense against cyber attacks.
Firewalls can be hardware, software. They work by inspecting data packets and determining whether to allow or block them based on a set of rules, examples:
- stop certain protocols from being used
- block data from or going to certain network addresses/website/IP adrdess
- stop attempts at **hacking**

- **hacking** the act of gaining unauthorised access to a computer system and the data it contains.
- **throughput** allow more data to pass through them.

#### Physical Security

**Physical security** ensures that critical parts of the network can only be physically accessed by authorised people. It might involve a burglar alarm, security tagging and physically locking down equipment, etc.\
The server room should be locked with an electronic lock system (better than traditional lock-and-key as entry/exit can be logged).\
With physical access to the server a criminal could more easily copy/modify/delete data or install malware that would allow them remote access.

- **Physical security** controlling access to critical parts of a network using physical methods (such as locked doors) rather than software.
- **malware** short for 'malicious software'. It is used as a generic term for any kind of software that is designed to disrupt the use of a computer system.

### Cloud Storage and Security

Security advantages:
- Data Encryption: Sensitive data is encrypted both at rest and in transit, making it unreadable to unauthorized users even if intercepted.
- Redundancy and Disaster Recovery: Cloud providers typically replicate data across multiple geographically dispersed data centers, ensuring data availability even in case of a local disaster. 
- Regular Security Updates: Cloud providers continuously update their security infrastructure to address emerging threats and vulnerabilies

Security disadvantages
- Vendor Dependency: Data security is heavily reliant on the chosen cloud provider's security practices and commitment to data protection. 
- Data Breach Risk: If a cloud provider experiences a data breach, sensitive user data could be compromised. 
- Misconfiguration Issues: Improper configuration of cloud storage settings by users can create security vulnerabilities. 
- Third-Party Access: Applications or services that access cloud storage may introduce additional security risks. 
- Compliance Concerns: Depending on the industry, meeting data privacy regulations with cloud storage can be challenging.

### Cyber Attacks

A **cyberr attack** is any kind of electronic attack on a computer system, server, network or other IT device in order to:
- gain access to data
- delete or modify information
- make the system unavailable for use
- physically damage a device connected to the network

### Types of cyber Attack

Most cyber attacks can be classed as exploiting either human behaviour or technical weaknesses.

- **cyber attack** any kind of malicious attack on a network-connected device
- **social engineering** any kind of attack on a computer system or network that takes advantage of how people behave and respond to certain situations
- **phishing** attempting to get confidential information by sending a user a message that appears to come from a trustworthy person or organisation.
- **shoulder surfing** gaining access to information by watching someone enter it into a computer system.
- **pharming** directing internet users to a fake website that looks like a real one, in order to obtain personal information such as passwords, accout numbers, etc.

### Social Engineering

Essentially, it's the act of tricking people into giving away personal data like passwords or financial details through human interaction. Three common forms of social engineering are **phishing**, **shoulder surfing** and **pharming**.

#### Phishing

A **phishing** attack is a cybercrime where a hacker sends a fraudulent communication (email, sms, phone conversation etc.) that appears to come from a reputable source, with the goal of stealing sensitive data or installing malware.

<img src="Img/U5_22_Phissing.png" width="500">

#### Shoulder Surfing

A shoulder surfing attack is where an attacker physically views a victim's device screen and keypad to steal personal information

#### Pharming

A pharming attack is a cyberattack that redirects a website's traffic to a fake site. It's done by installing a malicious program on a victim's computer to gain access to it. 

Pharming attacks can be carried out by:
- Changing a victim's computer's hosts file
- Exploiting a vulnerability in DNS server software

<img src="Img/U5_22_Pharming.jpg" width="500">

To prevent pharming:
- check that the http address of the site is the one you intended to visit
- check that there is a secure connection (https)
- check the site's security certificate
- install antivirus software

### Technical Weakness

#### Unpatched Software

The maker of the software will normally provide updates (patches) to fix security issuess as they are found. The patches to fix the security issues often have to be manually installed by a technician. Sometimes these patches get forgotten and the software remains vulnerable.

- **unpatched software** software that hasn't had the latest security updates applied to it, making it vulnerable to attack.

#### Usb Devices

USB device can potentially be a security threat because it might contain malware that could be transferred to your system or copy data to the attacker via the Internet.

#### Eavesdropping

In computer security, eavesdropping is the unauthorized interception of communication or data in transit between two devices. It can also refer to listening to or recording private conversations.

### Protecting Against Security Weaknesses

#### Design and Implementation of Software

The software and system designers will need to consider the following:
- What kind of authentication is needed?
- Do different users need different levels of access?
- Should warnings be issued before allowing users to carry out potentially risky operations?
- Does stored data need to be encrypted?
- What threats will the software face?


One major issue at this point is guarding against bad programming practice. Some programmers write poor quality code and don't consider how safe or secure their code is.

Another major issue is that some methods used in programming might lead to **code vulnerability**.

- **code vulnerability** a computer program (the code) that has been written in such a way that it creates a security issue that may be taken advantage of to gain access to the computer system or data within it.

Issues like these can be minimised by having regular code reviews. There are two main types of code review:
- Review by another programmer, usually someone who is more senior and has more experience.
- An automated review. The specialised software will highlight potential issues such as common vulnerabilities.

**Modular testing** is a software testing method that uses small, independent test cases to testing an application.

- **Modular testing** testing each block of code as it is completed to ensure the code works as expected.

### Other Security Measures to Protect From Cyber Attacks

- **audit trail** a record of activities that have taken place on a computer system. This record is generated  automatically and will record what has happened and who or what made the change.

- Use an **audit trail**, a log file that records all the activities on a computer system. Audit trails attack allow technicians to figure out what happened during a cyber attack.
- Use secure operating systems. Different operating systems are designed for different purposes.
- Provide effective network security. Well-educated technical staff with up-to-date training, implement policies that reduce the change of an attack, train users in best practice...

### Identifying Vulnerabilities

Ethical hacking is the practice of using the same tools and techniques as criminal hackers to identify and fix security vulnerabilities in a computer system, network, or application.

#### Penetration Testing

Penetration testing (pen testing) is where the IT systems of an organisation are deliberated attacked to find any weaknesses. Once the pen testing has beed completed, a report is presented explaining what issues were found.

#### Commercial Analysis Tools

It is possible to use software tools to scan vulnerabilities. They can be used to scan from within (internally) or from outside (externally).

#### Reviews of Network and User Policies

Reviews of Network and User Policies

All networks should have written policies that document:
- who is authorised to carry out various activities on the network
- how and when patches to software should be applied
- acces control
- password requirements, including how complex passwords should be and how often they should be changed.
- how security is set up and maintained on the network
- what data audit trails should collect and how long they should be kept for
- anything else relevant to the security and maintenance of the network.

All employees should sign the user policy. This user policy is likely to contain details of:
- what use of the network is allowed or not allowed
- what will happen to the user if they do something unacceptable
- how to report faults, problems and security issues
- security information, such as good practice when choosing and using passwords

Your computer or network is likely to be connected to the Internet using an **Internet Service Provider (ISP)**. 

The Internet uses the TCP/IP protocol stack to allow communication between all the different networks.

The Internet Protocol (IP) part of the stack provides the technical rules on addressing, provides each device or network with a unique address to send data to/from. Two versions of IP, IPv4 and IPv6 are currently in use.

The networks that are part of the Internet are linked together using **routers**.

When data are sent to another whose IP address it already knows the following happens:

1. THe sending computer splits the data into packets.
2. Each packet has a header that contains the sender's address, the destination address, the current packet number and the total number of packet that make up the data.
3. Each packet is sent to your ISP.
4. Your ISP will have a router. This router decide where to send the packet.
5. The packet is likely to end up at another router, which will again look at the destination IP address and forward it on.
6. Once a packet reaches its destination, the receiving computer will put the data back together from the packets.

#### How the World Wide Web works

Getting a web page or other file uing the World Wide Web involves a large number of steps that use different protocols and standards:

1. The user enter and URL (Uniform Ressource Locator) in the browser. A URL may be for a web page or an individual file.
2. The computer uses a system called the DNS server to find the IP address to the web server.
3. The web browser connects to the web server using the IP address and requests the relevant web page or other object.
4. A web page is transferred from one computer to another using HTTP or HTTPS.
5. Data sent from a web server to a web browser is in **HyperText Markup Language (HTML)**. The web browser displays the web page as described by the HTML.

- **HyperText Markup Language (HTML)** essentially a text document that contains any text to be displayed along with details how the text should be formatted, any hyperlinks and where they link to using URL, details of any objects such as picture or videos that should be shown.

Your computer or network is likely to be connected to the Internet using an **Internet Service Provider (ISP)**. 

The Internet uses the TCP/IP protocol stack to allow communication between all the different networks.

The Internet Protocol (IP) part of the stack provides the technical rules on addressing, provides each device or network with a unique address to send data to/from. Two versions of IP, IPv4 and IPv6 are currently in use.

The networks that are part of the Internet are linked together using **routers**.

When data are sent to another whose IP address it already knows the following happens:

1. THe sending computer splits the data into packets.
2. Each packet has a header that contains the sender's address, the destination address, the current packet number and the total number of packet that make up the data.
3. Each packet is sent to your ISP.
4. Your ISP will have a router. This router decide where to send the packet.
5. The packet is likely to end up at another router, which will again look at the destination IP address and forward it on.
6. Once a packet reaches its destination, the receiving computer will put the data back together from the packets.

### The World Wide Web

The World Wide Web (WWW) is a collection of web pages and other resources that are accessible over the internet. It's a system of interconnected documents that are linked together by hyperlinks, allowing users to navigate between them. The WWW is one of many applications that use the internet, but it's not the same as the internet itself.

#### Accessing the World Wide Web

You access the WWW using a program called a web browser. The web browser's job is to convert data received from a web server to a human-readable format.

#### How the World Wide Web works

Getting a web page or other file uing the World Wide Web involves a large number of steps that use different protocols and standards:

1. The user enter and URL (Uniform Ressource Locator) in the browser. A URL may be for a web page or an individual file.
2. The computer uses a system called the DNS server to find the IP address to the web server.
3. The web browser connects to the web server using the IP address and requests the relevant web page or other object.
4. A web page is transferred from one computer to another using HTTP or HTTPS.
5. Data sent from a web server to a web browser is in **HyperText Markup Language (HTML)**. The web browser displays the web page as described by the HTML.

- **HyperText Markup Language (HTML)** essentially a text document that contains any text to be displayed along with details how the text should be formatted, any hyperlinks and where they link to using URL, details of any objects such as picture or videos that should be shown.

### The need for IP addressing standards

An IP address is a set of numbers that are used to identify one particular computer.
IP addresses originally consisted of four 8-bit numbers for version 4 addresses (IPv4). IPv6 was instroduced so that more addresses would be available. IPv6 used 128 bits. An IPv6 address consist of eight groups of hexadecimal numbers separated by colons.
