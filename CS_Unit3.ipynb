## 12 Binary

- **Binary** is a base-2 system using only two digits: 0 and 1.

#### Why Binary?

Binary is needed because of the way in which computers work. The processor contains billions of *transistors*. They have only two states: on or off; they either *transmit* an electric current or they do not.

- **Digital** information represented by certain fixed values. Billions and billiona of 0 and 1. 
- **Analogue** using signals or information represented by a quantity that is continuously variable.

### Representing Information

- **Binary digit (bit)** the smallest unit of data that is reprensented in a computer. It has a single binary values, either 1 or 0.

A single bit offers two possibilities (0, 1). \
Two bits give four possible combinations (00, 01, 10, 11) \
Three bits would give eight combinations (000, 001, 010, 011, 100, 101, 110, 111) \
Etc.

The number of possible combinations is equal to 2 to the power of the number of bits used: $ N = 2^n$

### Two's Complement

The result of any number added to its negative equivalent should be 0.

#### Finding the Two's Complement of a Binary Number

- Flip all the bits. 1 -> 0, 0 -> 1.
- Add 1 to the result.

|  | |  |  |  |  |  |  |  |
| :----------: | --: | :--: | :--: | :--: |  :--: | :--: | :--: | :--: |
| Number| 0 | 0 | 0 | 1 | 1 | 1 | 0 | 0 |
| Flip| 1 | 1 | 1 | 0 | 0 | 0 | 1 | 1 |
| Add 1| 1 | 1 | 1 | 0|  0 | 1 | 0 | 0 |

Thus +$28_{10} = 1 1 1 0 0 0 1 1_2$ \
and -$28_{10} = 1 1 1 0 0 1 0 0_2$

#### Arithmetic Shifts

## 14 Data Storage and Compression

To ensure that data can be stored and transmitted efficiently, it can be smaller.

### Data Storage

- **nibble**: 4 bits
- **decimal prefix**: multiplies a unit by powers of 10
- **binary prefix**: multiplies a unit by powers of 2

Let's consider a file size of 293 093 376 bits or 36 636 673 bytes.
With a **decimal prefix** (Ubuntu) the size would be 36.6 MB. And with a **binary prefix** (Windows) it would be 34.6 MB.

Why? because **decimal prefix** use kilo as $10^3 = 1000$ but **binary prefix** use kilo as $2^{10} = 1024$

### Data Compression

- **compression**: changing the format of a data so that the size of the file becomes smaller.
- **bandwith**: the amount of data that can be carried on a network in a given period of time.
- **lossless compression**: compressing a file in such a way that it can be decompressed without any loss of data.
- **lossy compression**: compression where some of the data is removed; the original file cannot be restored when the lossy file is decompressed.

File **compression** is very important. Compressed file use les bandwith to upload and download!
Compression algorithms are used to make the files as small as possible. There are two types of compression - **lossless** and **lossy compression**.

### Lossless Compression

#### Run-Length Encoding (RLE)

Run-Length Encoding is used to reduce the size of a repeating string of items. The repeating string is called a rin and is represented by two bytes - the first byte represents the number of times the item of information is repeated and the second, the item of imformation.

for example:\

become:\
5S6Y2C2V1A3M6C3Y7I4Q

This method is only effective where there are strings of the same data. If the original data is very different, without repeating strings, then the encoded version would have agreater file size.

### Modern Methods of Data Encryption

#### Asymmetric Encryption

Every used has two keys, a public key and a private key. A message encrypted with a particular public key can only be decrypted by the corresponding private key.

#### Symmetric Encryption

Symmetric encryption algorithms (used by HTTPS) encrypt and decrypt a message using the same key. The client generates the key using agreed algorithm and sends it to the server using the server's public key (asymmetric encryption).
