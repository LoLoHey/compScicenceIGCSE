## 12 Binary

- **Binary** is a base-2 system using only two digits: 0 and 1.

#### Why Binary?

Binary is needed because of the way in which computers work. The processor contains billions of *transistors*. They have only two states: on or off; they either *transmit* an electric current or they do not.

- **Digital** information represented by certain fixed values. Billions and billiona of 0 and 1. 
- **Analogue** using signals or information represented by a quantity that is continuously variable.

### Representing Information

- **Binary digit (bit)** the smallest unit of data that is reprensented in a computer. It has a single binary values, either 1 or 0.

A single bit offers two possibilities (0, 1). \
Two bits give four possible combinations (00, 01, 10, 11) \
Three bits would give eight combinations (000, 001, 010, 011, 100, 101, 110, 111) \
Etc.

The number of possible combinations is equal to 2 to the power of the number of bits used: $ N = 2^n$

### Two's Complement

The result of any number added to its negative equivalent should be 0.

#### Finding the Two's Complement of a Binary Number

- Flip all the bits. 1 -> 0, 0 -> 1.
- Add 1 to the result.

|  | |  |  |  |  |  |  |  |
| :----------: | --: | :--: | :--: | :--: |  :--: | :--: | :--: | :--: |
| Number| 0 | 0 | 0 | 1 | 1 | 1 | 0 | 0 |
| Flip| 1 | 1 | 1 | 0 | 0 | 0 | 1 | 1 |
| Add 1| 1 | 1 | 1 | 0|  0 | 1 | 0 | 0 |

Thus +$28_{10} = 1 1 1 0 0 0 1 1_2$ \
and -$28_{10} = 1 1 1 0 0 1 0 0_2$

#### Finding Two's Complement Numbers back into Denary

##### Method 1, Reverse Calculation

The result is denary 28.

### Binary Shifts

In denary , when we want to multiply a number by 10, we move each number one place value to its left and add a 0  to the end.\
In binary left and right, **binary left** can be used for multiplication and division by powers of 2.

- **binary shift:** an operation done on all the bits of a binary value in which they are moved by a specific number of places to either the left or right.

#### Logical Shifts
For unsigned numbers, logical shifts are performed in the same way as in denary.

| $20_{10}$ | 0 | 0 | 0 | 1 | 0 | 1 |0 | 0 | |
| $20 / 2^2$ | **0**| **0** | 0 | 0 | 0 | 1 |0 | 1 | =5 |

A logical shift on integers can make a number less precise.

#### Arithmetic Shifts

- **Arithmetic shift:** used for signed binary numbers. When performing a right shift, the bits at the left are replaced by copies of the most significant bit.

##### Left Arithmetic Shift

The left-most-bit (MSB) is not included because it must remain in place to indicate the sign.

|  | |  |  |  |  |  |  |  | |

| -28*2| 1 | 1 | 0 | 0| 1 | 0 | 0 | **0** | = -56 |

##### Right Arithmetic Shift

The bits are shifted to the right. They are replaced at the left by copies of the MSB.

### Representation of Bitmap Images

Of course, images are also represented and stored as binary code. The basic unit of a bitmap image is the **pixel**.

- **pixel**: short for 'picture element', the smallest single point of colour in a graphic image.

#### Image Size and Resolution

Image size 640 x 480: First number indicates the numbers of pixels along the width of the image and the second is the number in the height.

- **resolution**: the number of pixels per inch when the image is displayed.

### Recording Sound Waves

#### Analogue Recording

Analogie recording represents the continuous changes in ai pressure caused by sound waves, as changes in voltage.

#### Digital Recording

Computer cannot represent a continuous change. Sound data must be represented as stream of 1s and 0s. The sound is sampled.

- **Sampling**: Taking measurments of a sound wave at regular but different intervals of time. 
- **Sample Rate**: Number of sample per second, (e.g. 44.1 kHz)
- **Bit Depth**: The number of bits used to encore the information from each sample. Increasing the bit depth used increases the amount of detail.

#### Fidelity

The fidelity is influcenced by the saple rate and the bit depth.

- Cd: 44.1 Khz with 16 bits
- Blue Ray: 96 Khz up to 24 bits

One important detail is the *dynamic* range of the sound - The range of volumes of sound in the music.

### Digital Audio File Sizes

The size of a digital audio sound depends on the following:

- dample rate
- bit depth
- duration
- number of channels

#### Calculating file size

file size in bits = sample rate * bit depth * duration (in s) * number of channels
