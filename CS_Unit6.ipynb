## 24 Computing and the Environmental Impact of Technology

Everythin we do has an impact on the environment and manufacture, use and disposal computing devices is no exception.

The more **computing technology** we buy, the more we throw away. We are using up resources of non-renewable materials, creating potentially harmful **e-waste**, consuming energy and damaging people's health.

- **computing technology** and all-encompassing term referring to the hardware, software and infrastructure that underpin current and emerging computer systems.
- **e-waste** any form of discarded electronic equipment, including computing technology.

### Manufacture

#### Raw Material Extraction

A number of non-renewable natural ressources are used in the manufacture of computer product: sand, oil, various metal...

Some materials are radioactive and can contaminate the environment.

Mining of raw materials can also be very damaging for the local environment if not control properly.

#### Production

The production of computers uses a lot of energy and also uses a very large amount of water.

Governments are also imposing measures to promote material recycling.

### Usage

A extremely vast quantity of energy is required everyday to power the bilions of computer like devide in the world.
In recent years, considerable efforts have been made to improce tthe energy efficiency of computing device.

Data centers are major energy users. Vast amounts of electricity are needed to power and cool all the computer equipment that is needed. Energy efficiency and the use of renewable energy can significantly reduce the **carbon footprint**.

- **carbon footprint** the amount of carbon dioxide an individual or organisation produces as a result of the energy they consume

### Disposal

The primary problem with disposing of e-waste is that electronic devices often contain hazardous chemicals like lead, mercury, cadmium, and brominated flame retardants, which can severely contaminate the environment and harm human health if not disposed of properly, potentially polluting soil, water, and air when dumped in landfills or improperly recycled.

## 25 Privacy

An action might be legal, but not necessarily **ethical**. 
Computing technology confers a wide range of social and economic benefits, but it also createa challenging ethical issues. **Privacy** and security are two of them.

- **ethics** a set of moral principles that govern a person's behavior
- **privacy** the right to be left alone and free from unwanted scrutiny and intrusion
- **personal data** information that is personal and unique to an individual

### Personal Data

Each time you are active online you are adding to a store of **personal data** that is held about you. This personal data is stored on servers that do not belong to you, but Facebook, Google etc.

Weak security could result in personal information falling into the wrong hands. Resulting to vulnerability to phishing attacks, scams,**indentity theft** and fraud.

**Personal data** allows organisation to provide personalised services. These can be benefitial but have also drawbacks.

HTTP cookies (also called web cookies, Internet cookies, browser cookies, or simply cookies) are small blocks of data created by a web server while a user is browsing a website and placed on the user's computer or other device by the user's web browser. 

- **indentity theft** the stealing of another person's personal details, such as their bank account number, sort code or passport number, for the purpose of making purchases and running up debts in their name

#### Big Data

Big data refers to extremely large and diverse collections of structured, unstructured, and semi-structured data that continues to grow exponentially over time.

Big data offers significant benefits like improved decision making, better customer understanding, operational efficiency, and trend prediction by analyzing large volumes of data; however, it also presents challenges including data privacy concerns, complex storage and processing needs.

#### Surveillance

**Surveillance technology** is everywhere. How many times per day you are recorded on CCTV. Soon drones might hover you. Maybe your parent are tracking your phones?!.

Many **Whiste-blowers** raised awareness of the extent to which governements worldwide are now monitoring and spying on their citizens.

- **Surveillance technology** CCTV, drones, number plate recognition, bugging and tracking devices used to monitor and record people's activities, often without their knowledge
- **Whistle-blower** someone who draws attention to the activities of an organisation or person believed to be actnig illegally or unethically.

#### Location-Based Services

A location-based service (LBS) is a software service that uses a user's location to provide information or services. LBSs use a user's device's location data to provide relevant information about their current location.

Examples of LBSs
- Maps and navigation: Maps, routing, and assisted navigation services
 - Tracking services: Real-time traffic updates, vehicle tracking, and friends and family finder

**location-based-services** services that enable people to access and share real-time loation information online.

## 26 Digital Inclusion

**Digital inclusion** refers to the active efforts to ensure everyone has access to and can effectively use digital technologies, like the internet, while **digital divide** describes the gap between those who have access to these technologies and those who do not, highlighting the inequality in digital participation across different groups of people

- **digital inclusion** ensuring that everyone has affordable access to computing technology and the necessary skills to take advantage of it.
- **digital divide** the gap between people who are technology-empowered and those who are technology-excluded.

## 28 Computing and the Legal Impact of Technology

### Intellectual Property

An IP is a unique creative product of a human mind. A piece of software, a computer game, a desgin for a new processor, a digital image, a piece of music and a literary work are all examples of IP.

- **Intellectual property (IP)** a creation of the human min that is unique and has a commercial value

#### Copyright and Patents

Copyright only protects the expression of an idea, not the idea itself. A **patent** offers more protection than copyright. It protects the idea or design of an invention, rather than just a particular form of it.

- **Patent** an exclusive right granted to an inventor to make, use and sell an invention for a fixed period of time.

### Licensing

A software license is a legal agreement that gives users permission to use software. It also outlines the rights and responsibilities of the software's creator, provider, and user.
Before you can install the software, you have to agree to the terms of its licence.

- how many copies of the software you are alloed to use
- whether you can install the software on more than one computer
- what type of organisation can use the software
- how long the software can be used for.

If a computer scientist wants to permit other people to use their code without charge, they can use an open-source licence to specify what restriction (if any) there are. For example, **Creative Commons** in UK provides a way for the creators to share their work.

- **Creative Commons** an organisation that allows people to set copyright terms for their intellectual property. One use of a Creative Commons licence is to allow pwople to copy material as long as it is not used commercially.

### Open-source and proprietary software

**Open-source software** is freely available on the Internet. Anyone is permited to edit the code and pass it on others, providing they don't charge a fee.

| Advantages | Disadvantages|
| ---------- | ------------ |
| It is free to use | It might not be particularly user friendly |
| It can be modified | There might be little or no technical support available |
| It can be used to demonstrate programming concepts | Criminals may be able to identify and exploit vulnerabilities in the code |

**Proprietary software** os the opposite of opens source - it is closed source. This means its source code is protected and users are not allowed to modify it.

- **open-source software** software that is free to edit and redistribute
- **proprietary software** software that belongs to an individual or company. Its licence specifies that users are not allowed to modify the source code and places restrictions on its use.

It is illegal for anyone to gain access to computer systems without permission.
